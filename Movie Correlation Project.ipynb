```python
# import libraries

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)# Adjust the configuration of the plots that will create

# Read data

df = pd.read_csv('/Users/../movies.csv')
```

```python
# Let's look at the data

df.head()
```

```python
# check if there is any missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# Remove n/a cell

df = df.dropna()
```

```python
# Confirm if any blank cell is removed

df.isnull().sum()
```

```python
# Data Types for our columns

print(df.dtypes)
```

```python
# Change data type from float to integer

df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')

df['votes'] = df['votes'].astype('int64')
```

```python
df.head()
```

```python
# Create correct year column

df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)
```

```python
df.head()
```

```python
df = df.sort_values(by = ['gross'],inplace = False, ascending = False)
```

```python
pd.set_option('display.max_rows', None)
```

```python
df['company'].sort_values(ascending = False)
```

```python
# Hypothesis: 
# Budget has high correlation
# Company has high correlation
```

```python
# Scatter plot with budget vs gross

plt.scatter(x = df['budget'], y = df['gross'])

plt.title('Budget vs Gross Earnings')

plt.xlabel('Gross Earnings')

plt.ylabel('Budget')

plt.show()
```

```python
# Plot the budget vs gross using seaborn

sns.regplot( x = 'budget', y = 'gross', data = df, scatter_kws = {"color" : "red"}, line_kws = {"color" : "blue"})
```

```python
df.corr() # default method is pearson (pearson/kendall/spearman)
```

```python
# By examining by graphic and correlation calculate, we can see there is a high correlation between budget and gross.
```

```python
correlation_matrix = df.corr() 

sns.heatmap(correlation_matrix, annot = True)

plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movie Faetures')

plt.ylabel('Movie Faetures')

plt.show()
```

```python
# Let's examine the correlation of company and gross
```

```python
# Convert data type object to category

df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized.head()
```

```python
correlation_matrix = df_numerized.corr() 

sns.heatmap(correlation_matrix, annot = True)

plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movie Faetures')

plt.ylabel('Movie Faetures')

plt.show()
```

```python
df_numerized.corr()
```

```python
correlation_matrix = df_numerized.corr()

corr_pairs = correlation_matrix.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs)>0.5]

high_corr 
```
