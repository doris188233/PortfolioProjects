<h4> Exploratory Data Analysis

- <h5> What is the overall sales?

```python
df['order_date'].min()
```

```python
df['order_date'].max()
```

```python
# Retrive the month and year from order_date column

df['month_year'] = df['order_date'].apply(lambda x: x.strftime('%Y-%m'))
```

```python
# sum up the sales order by month, only display the sales and set index number

df_trend = df.groupby('month_year').sum()['sales'].reset_index()
```

```python
# create plot graph with y axis is sales and x asix is month  

plt.figure(figsize = (15,6))

plt.plot(df_trend['month_year'], df_trend['sales'], color = 'red')

plt.xticks(rotation="vertical", size=8)

plt.xlabel('Months in Year')

plt.ylabel('Total Sales')

plt.show()
```

- <h5> Which are the top 10 best selling items in sales revenue?

```python
prod_salse = pd.DataFrame(df.groupby('product_name').sum()['sales'])
```

```python
prod_salse = prod_salse.sort_values('sales', ascending = False)
```

```python
prod_salse[:10]
```

- <h5> Which is the top 3 best selling product in term of number of sold?

```python
quantity_sold_prod = pd.DataFrame(df.groupby('product_name').sum()['quantity'])
```

```python
Top_quantity_sold_prod = quantity_sold_prod.sort_values('quantity', ascending = False)
```
